# Part 3: Some single variable plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
```

Load dataset and do preprocessing.

```python
data = pd.read_csv('../../incident_reports.csv', parse_dates=['Date'])
```

```python
def getHour(time):
    return int(time.split(':')[0])

def getMinute(time):
    return int(time.split(':')[1])

data['Hour'] = data.Time.map(getHour)
data['Minute'] = data.Time.map(getMinute)
```

**First create a jitter-plot (that is, code up something like Figure 2-1 from DAOST from scratch), but based on SF Police data. My hunch from inspecting the file is that the police-folks might be a little bit lazy in noting down the exact time down to the second. So choose a crime-type and a suitable time interval (somewhere between a month and 6 months depending on the crime-type) and create a jitter plot of the arrest times during a single hour (e.g. 13-14).**

We have chosen to focus on assaults in January to May between 13 and 14 of each day.

```python
assaults = data[data['Category'] == 'ASSAULT']
assaults_filtered = assaults[(assaults.Date.dt.month < 6) & (assaults.Hour == 13)]
assaults_by_minute = assaults_filtered.groupby('Minute')['IncidntNum'].count()
```

```python
x_spread = 0
y_spread = 1
plt.figure(figsize=(15,7))
for i in range(60):
    x_noise = np.random.rand(assaults_by_minute[i])
    y_noise = np.random.rand(assaults_by_minute[i])
    x = i + x_noise * (2 * x_spread) - x_spread
    y = y_noise * y_spread
    plt.scatter(x, y, marker='o', edgecolors='black', facecolors='none')
plt.xlabel('Minute of the hour between 13 and 14')
plt.show()
```

We see that assaults are mostly reported every 5th and 10th minute. Especially the 0th and 30th minute marks seem to have a high concentration of incidents. Indeed the hunch holds true for the selected date and time interval. It therefore appears that the hunch is correct.

**Now it's time for histograms. We will now create two versions of Figure 2-2.**

**Pick two crime-types with different geographical patterns and a suitable time-interval for each (you want between 1000 and 10000 points in your histogram)**

We have selected drug/narcotics and prostitution throughout January to May and between midnight and 7 in the morning.

```python
filtered = data[(data.Date.dt.month < 6) & ((data.Hour > 0) & (data.Hour < 7))]
drug = filtered[filtered['Category'] == 'DRUG/NARCOTIC']
prost = filtered[filtered['Category'] == 'PROSTITUTION']
```

Let's check that each crime-type has between 1000 and 10000 points in the specified time-interval.

```python
print(f'No. of drugs/narcotics incidents: {len(drug)}')
print(f'No. of prostitution incidents: {len(prost)}')
```

**Then take the latitude part of the GPS coordinates for each crime and bin the latitudes so that you have around 50 bins across the city of SF. You can use your favorite method for binning. I like numpy.histogram -- that gives you the counts and then you do your own plotting.**

```python
def histogram(series, max_val=None, min_val=None, no_of_bins=50, **kwargs):
    if max_val is None:
        max_val = series.max()
    if min_val is None:
        min_val = series.min()
    bins = np.linspace(min_val, max_val, no_of_bins)
    plt.hist(series, bins=bins, histtype='step', **kwargs)
```

We will plot the histograms together to compare the lattitudes of the two crime-types.

```python
plt.figure(figsize=(10, 5))
histogram(drug.Y, label='Drugs/narcotics')
histogram(prost.Y, label='Prostitution')
plt.legend(frameon=False, loc='upper left')
plt.xlabel('Lattitude')
plt.ylabel('No. of incidents')
plt.show()
```

This shows that drugs/narcotics and prostituion happen close to it each other (when we don't take longitude into account) but they stil don't happen at the exact same locations. Interestingly both distributions seem to have two spikes which suggests that each of the two crime-types mostly happen in two city districts. Another interesting finding is that drug/narcotics has a long tail (to its left) while prostitution is mostly flat outside its spikes. This suggests that drug/narcotics incidents happens throughout the entire city while prostitution is more concentrated to a few areas.

**Next up is recreating Figure 2-4 from DAOST, but with the data you used to create Figure 2-1. To create the kernel density plot, you can either use gaussian_kde from scipy.stats (for an example, check out this stackoverflow post) or you can use seaborn.kdeplot.**

```python
plt.figure(figsize=(10, 5))
histogram(assaults_filtered.Minute, no_of_bins=60, density=True, label='Histogram')
sns.kdeplot(assaults_filtered.Minute, bw=1, label='KDE, bandwidth = 1')
sns.kdeplot(assaults_filtered.Minute, bw=2, label='KDE, bandwidth = 2')
plt.xlabel('Minute of the hour 13-14')
plt.ylabel('Density')
plt.xlim(0, 60)
plt.title('Density of assaults in the hour 13-14 throughout January to May')
plt.show()
```

It is clear from this plot that the KDEs need to have the right bandwidth to represent the true density correctly. In this case a bandwidth of 1 seems to work fairly well. However, a bandwidth of 2 grossly underestimates the true density at the spikes while overestimating outside of the steps. This gives an impression that the distribution is a lot flatter than it actually is.

**Now grab 25 random timepoints from the dataset you've just plotted and create the same Figure 2-4 plot once again, but with fewer points. Does this shed light on why I think KDEs can bee misleading?**

```python
samples_indices = np.random.choice(range(len(assaults_filtered)), size=25, replace=False)
```

```python
assaults_samples = assaults_filtered.iloc[samples_indices]
```

```python
plt.figure(figsize=(10, 5))
histogram(assaults_samples.Minute, no_of_bins=60, density=True, label='Histogram')
sns.kdeplot(assaults_samples.Minute, bw=1, label='KDE, bandwidth = 1')
sns.kdeplot(assaults_samples.Minute, bw=2, label='KDE, bandwidth = 2')
plt.xlabel('Minute of the hour 13-14')
plt.ylabel('Density')
plt.xlim(0, 60)
plt.title('Density of a sample of assaults in the hour 13-14 throughout January to May')
plt.show()
```

It is hard to compare the KDEs with the previous unsampled versions by scrolling up and down between the two plots. Therefore we have created another plot to directly compare the KDEs for the unsampled and samples data (bandwidth = 1).

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(assaults_filtered.Minute, bw=1, label='Original - KDE, bandwidth = 1')
sns.kdeplot(assaults_samples.Minute, bw=1, label='Sample - KDE, bandwidth = 1')
plt.xlabel('Minute of the hour 13-14')
plt.ylabel('Density')
plt.xlim(0, 60)
plt.title('Comparison of assaults and samples\' KDEs width bandwidth = 1')
plt.show()
```

It is clear from this plot that the KDEs of the orignal and sampled data are vastly different. They both have a spike around the 30 minute mark but, besides from that, they represent two different distributions. We know that the orange KDE comes from a small sample of 25 observations but this information is not possible to observe from the KDE which makes it appear as if both KDEs are estimated from a sufficient amount of data and therefore KDEs can be misleading.

**Now box plots. Here, I'd like you to use the box plots to visualize fluctuations of how many crimes happen per day. For the 15 focus crimes defined last week.**

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

**For the full time-span of the data, calulate the number of crimes per day within each category for the entire duration of the data (this should result in 15 datasets, on for each focus crime).**

```python
def get_category_date_count(category):
    return data[data.Category == category].groupby('Date')['IncidntNum'].count().to_numpy()

boxplot_data = list(map(get_category_date_count, focuscrimes))
```

**Create a box-and whiskers plot showing the mean, median, quantiles, etc for each of the 15 crime-types and show them side-by-side. There are many ways to do this, I like to use matplotlibs's built in functionality**

```python
plt.figure(figsize=(14, 7))
plt.boxplot(boxplot_data, labels=focuscrimes, vert=False)
plt.xlabel('Crimes per day')
plt.show()
```

**What does this plot reveal that you can't see in the plots from last time?**

The boxplots show us the same order in the most frequent crimes as for the histograms of focus crimes. E.g. larcency/theft is more frequent than stolen property. In regard to what we can't see in the plots from last time, we also get the variation in each focus crime from the box-and-whiskers plot. An example is that there is a huge variation in larcency/theft as well as e.g. assault. Other focus crimes - such as "driving under the influence" - appear to have little to no variation.
