Define a function to filter the data from the original dataset:

```python
def filter_crimes(data, category, year, month, before_day):
    crimes = data[(data['Category'] == category)]
    crimes = crimes[(crimes['Date'].dt.year == year) & (crimes['Date'].dt.month.isin(month)) & (crimes['Date'].dt.day < before_day)]
    return crimes
```

**Comment on the differences. What can you see using the scatter-plots that you can't see using the heatmaps? And vice versa: what does the heatmaps help you see that's difficult to distinguish in the scatter-plots?**

The drugs/narcotic crime is very much limited to a specific area and has few high density places within that.
The larceny/theft crime, on the other hand, covers a wider area and has more high density regions. A couple in the same area as the drugs/narcotic one, and others more north and west.
From the heatmap, it looks like theft is a more common crime and affecting a wider area of San Francisco, whereas the drug crimes are less and limited to specific places.

Heatmaps are very intiutive tool to detect density differences in different areas and to get a direct visual feedback on how vast a crime affected area is and where the hot points are. This cannot be achieved with the same level of intuitiveness with scatter plots. 
Scatter plots, on the other hand, give a realistic idea of how many occurrences happened in each zone by plotting each occurrence. This allows to have a high degree of details and fidelity over the position and amount of crimes happened. Heatmaps however hide this sort of details when plotting densities. 

Scatter plot:
* **Pros:**
    * This plot allows to see all the occurrences spread on the plotted area. This means that it gives a realistic idea of how many occurrences happened in each zone. 
* **Cons:**
    * It is difficult to get a precise idea of how the distribution and concentration look like, over the area. High density areas tend to look like an agglomerate of color, because of all the circles in the same area. It is though difficult to distinguish the different agglomerate in terms of which ones are more dense than others.

Heatmaps:
* **Pros:**
    * These maps highlight the concentration of crimes in specific areas and the distribution of the different densities over the map. It is then easy to compare the differents zones' densities just by comparing their color. 
* **Cons:**
    * Hide away the real number of occurrences. It is not easy to determine how many crimes a high density zone contain.


**Comment on the effect on the various parameters for the heatmaps. How do they change the picture?**

From the official documentation, these are the parameter which can be passed to the heatmap:
* data (list of points of the form [lat, lng] or [lat, lng, weight]) – The points you want to plot. You can also provide a numpy.array of shape (n,2) or (n,3).
* name (string, default None) – The name of the Layer, as it will appear in LayerControls.
* min_opacity (default 1.) – The minimum opacity the heat will start at.
* max_zoom (default 18) – Zoom level where the points reach maximum intensity (as intensity scales with zoom), equals  maxZoom of the map by default
* max_val (float, default 1.) – Maximum point intensity
* radius (int, default 25) – Radius of each “point” of the heatmap
* blur (int, default 15) – Amount of blur
* gradient (dict, default None) – Color gradient config. e.g. {0.4: ‘blue’, 0.65: ‘lime’, 1: ‘red’}
* overlay (bool, default True) – Adds the layer as an optional overlay (True) or the base layer (False).
* control (bool, default True) – Whether the Layer will be included in LayerControls.
* show (bool, default True) – Whether the layer will be shown on opening (only for overlays).

We will not comment on the effect of all the parameters since many of them appear to be self-explanatory. Instead we have chosen to comment on the following parameters:
* radius
* blur
* min_opacity

#### Radius and Blur parameter

By reducing both the radius and the blurriness, it is possible to achieve a heatmap which recalls the scatter plot one: with sharp circles for each occurrence. The color of the circles, however, still reppresent the concentration of crimes, allowing us to have more information than a scatter plot.

```python
def getHeatMap(data):
    m = folium.Map([37.77919, -122.41914], tiles='stamentoner', zoom_start=13)
    coordinates = data[['Y', 'X']].values
    m.add_child(plugins.HeatMap(coordinates, radius=10, blur=0.1))
    return m
```

#### Blur parameter

This parameter determines how blurry the circles will be on the map. Decreasing it allows to draw sharper circles. By reducing the blurrines, the achieved effect is to a more precise reppresentation of the occurrences, more similar to a scatter plot.

```python
def getHeatMap(data):
    m = folium.Map([37.77919, -122.41914], tiles='stamentoner', zoom_start=13)
    coordinates = data[['Y', 'X']].values
    m.add_child(plugins.HeatMap(coordinates, radius=10, blur=10))
    return m
```

**First create a jitter-plot (that is, code up something like Figure 2-1 from DAOST from scratch), but based on SF Police data. My hunch from inspecting the file is that the police-folks might be a little bit lazy in noting down the exact time down to the second. So choose a crime-type and a suitable time interval (somewhere between a month and 6 months depending on the crime-type) and create a jitter plot of the arrest times during a single hour (e.g. 13-14).**

We have chosen to focus on assaults in January to May between 13 and 14 of each day.

```python
assaults = data[data['Category'] == 'ASSAULT']
assaults_filtered = assaults[(assaults.Date.dt.month < 6) & (assaults.Hour == 13)]
assaults_by_minute = assaults_filtered.groupby('Minute')['IncidntNum'].count()
```

```python
x_spread = 0
y_spread = 1
plt.figure(figsize=(15,7))
for i in range(60):
    x_noise = np.random.rand(assaults_by_minute[i])
    y_noise = np.random.rand(assaults_by_minute[i])
    x = i + x_noise * (2 * x_spread) - x_spread
    y = y_noise * y_spread
    plt.scatter(x, y, marker='o', edgecolors='black', facecolors='none')
plt.xlabel('Minute of the hour between 13 and 14')
plt.show()
```

We see that assaults are mostly reported every 5th and 10th minute. Especially the 0th and 30th minute marks seem to have a high concentration of incidents. Indeed the hunch holds true for the selected date and time interval. It therefore appears that the hunch is correct.

**Then take the latitude part of the GPS coordinates for each crime and bin the latitudes so that you have around 50 bins across the city of SF. You can use your favorite method for binning. I like numpy.histogram -- that gives you the counts and then you do your own plotting.**

```python
def histogram(series, max_val=None, min_val=None, no_of_bins=50, **kwargs):
    if max_val is None:
        max_val = series.max()
    if min_val is None:
        min_val = series.min()
    bins = np.linspace(min_val, max_val, no_of_bins)
    plt.hist(series, bins=bins, histtype='step', **kwargs)
```

We will plot the histograms together to compare the lattitudes of the two crime-types.

```python
plt.figure(figsize=(10, 5))
histogram(drug.Y, label='Drugs/narcotics')
histogram(prost.Y, label='Prostitution')
plt.legend(frameon=False, loc='upper left')
plt.xlabel('Lattitude')
plt.ylabel('No. of incidents')
plt.show()
```

This shows that drugs/narcotics and prostituion happen close to it each other (when we don't take longitude into account) but they stil don't happen at the exact same locations. Interestingly both distributions seem to have two spikes which suggests that each of the two crime-types mostly happen in two city districts. Another interesting finding is that drug/narcotics has a long tail (to its left) while prostitution is mostly flat outside its spikes. This suggests that drug/narcotics incidents happens throughout the entire city while prostitution is more concentrated to a few areas.

**Next up is recreating Figure 2-4 from DAOST, but with the data you used to create Figure 2-1. To create the kernel density plot, you can either use gaussian_kde from scipy.stats (for an example, check out this stackoverflow post) or you can use seaborn.kdeplot.**

```python
plt.figure(figsize=(10, 5))
histogram(assaults_filtered.Minute, no_of_bins=60, density=True, label='Histogram')
sns.kdeplot(assaults_filtered.Minute, bw=1, label='KDE, bandwidth = 1')
sns.kdeplot(assaults_filtered.Minute, bw=2, label='KDE, bandwidth = 2')
plt.xlabel('Minute of the hour 13-14')
plt.ylabel('Density')
plt.xlim(0, 60)
plt.title('Density of assaults in the hour 13-14 throughout January to May')
plt.show()
```

It is clear from this plot that the KDEs need to have the right bandwidth to represent the true density correctly. In this case a bandwidth of 1 seems to work fairly well. However, a bandwidth of 2 grossly underestimates the true density at the spikes while overestimating outside of the steps. This gives an impression that the distribution is a lot flatter than it actually is.

```python
vehicle_data = data[data["Category"] == "VEHICLE THEFT"][["Dst", "Day", "Hour", "Month"]]
vehicle_clean = vehicle_data.sample(frac=1).reset_index(drop=True)
vehicle_train = vehicle_clean[:10000]
vehicle_test = vehicle_clean[10000:11000]

drug_data = data[data["Category"] == "DRUG/NARCOTIC"][["Dst", "Day", "Hour", "Month"]]
drug_clean = drug_data.sample(frac=1).reset_index(drop=True)
drug_train = drug_clean[:10000]
drug_test = drug_clean[10000:11000]

print(vehicle_train.head())
print(drug_train.head())
```

The rows are shuffled, and the dataframes are then split into train and test data -  with size of 10.000 and 1.000 respectively.

**Then we are going to to build is a decision tree (or a random forest) that takes as input the four labels (Hour-of-the-day, Day-of-the-week, Month-of-the-year, and PD-District) of some crime (from one of the two categories) and then tries to predict which category that crime is from.**

**Some notes/hints**

- **A good option is the DecisionTreeClassifier.**
- **Since you have created a balanced dataset, the baseline performance (random guess) is 50%. How good can your classifier get?**

Now we are ready to train our first model.

```python
X = vehicle_train.append(drug_train)
Y1 = pd.Series(0, index=list(range(10000)))
Y2 = pd.Series(1, index=list(range(10000)))
Y = Y1.append(Y2)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
def stats(clf, vehicle_testset, drug_testset):
    n = len(vehicle_testset) + len(drug_testset)
    v_predictions = clf.predict(vehicle_testset)
    d_predictions = clf.predict(drug_testset)
    accuracy = ((v_predictions == 0).sum() + (d_predictions == 1).sum()) / n
    return accuracy
```

```python
stats(clf, vehicle_test, drug_test)
```

With an accuracy around 65% we are not much better than the 50% baseline. This could be improved using different approaches (more data, regularization etc.). We will attempt to improve it with a random forest classifier.
